```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA,FastICA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.feature_selection import SelectFromModel
import time
```

```python
# read column names from excel
column_names_excel= pd.read_excel(r'C:\Users\meghs\Documents\Big data\column_names.xlsx', 'Sheet2',header=None)
column_names=column_names_excel.values.tolist()  

#read data, replace ? with NaN
df = pd.read_csv(
    filepath_or_buffer=r'C:\Users\meghs\Documents\Big data\arrhythmia.data', 
    header=None,
    sep=',',na_values='?')

#name columns
df.columns = column_names  #renamed all columns

#data with NaN replaced by 0, and all 279 features
X_data_all=df.iloc[:,0:279].fillna(0)
```

```python
#cleaning data
X_data_o=df.iloc[:,0:279]
X_data_n=X_data_o.dropna(axis='columns') #drop NaN columns
X_data=X_data_n.loc[:, X_data_n.std()>0] #keep columns if standard deviation is greater than zero else delete the column

print("\n Shape of original data: ",X_data_o.shape,
      "\n Shape of data after NaN columns are dropped: ",X_data_n.shape,"\t Dimesion reduced by:",np.subtract(X_data_o.shape, X_data_n.shape),
      "\n Shape after same variance Columns are dropped: ",X_data.shape,"\t Dimesion reduced by:",np.subtract(X_data_o.shape, X_data.shape))


#target data
y_target=df.iloc[:,-1]

#plot instances in each class
pd.value_counts(y_target).plot(kind="bar")
plt.title('Number of instances for each class')
```

```python
#split data
X_data1_o=X_data_all[0:220]
y_target1=y_target[0:220]
X_data2_o=X_data_all[220:452]
y_target2= y_target[220:452]

#knn for original dimesion
X_train_o, X_test_o, y_train_o, y_test_o= train_test_split(X_data1_o,y_target1,test_size=0.2, random_state=0)

# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 15)
train_accuracy_o = np.empty(len(neighbors))
#train_error_o = np.empty(len(neighbors))
test_accuracy_o= np.empty(len(neighbors))
#test_error_o = np.empty(len(neighbors))
test_accuracy2_o = np.empty(len(neighbors))

time_o=[]

# Loop over different values of k
for i, k in enumerate(neighbors):
    # Setup a k-NN Classifier with k neighbors: knn
    knn_o = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn_o.fit(X_train_o,y_train_o)
    
    
    #Compute accuracy on the training set
    
    train_accuracy_o[i] = knn_o.score(X_train_o, y_train_o)
    #train_error_o[i] =(1-knn_o.score(X_train_o, y_train_o))*100

    #Compute accuracy on the testing set
    knn_o.predict(X_test_o)
    test_accuracy_o[i] = knn_o.score(X_test_o, y_test_o)
    #test_error_o[i] =(1-knn_o.score(X_test_o, y_test_o))*100
    
    knn_o.predict(X_data2_o)
    test_accuracy2_o[i] = knn_o.score(X_data2_o, y_target2)

# Generate plot
print('\n Training Accuracy \n',train_accuracy_o,'\n \n Testing Accuracy \n',test_accuracy_o,'\n \n Testing Accuracy on new data \n',test_accuracy2_o)
plt.title('k-NN: Varying Number of Neighbors for original dimensions')
plt.plot(neighbors, test_accuracy_o, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy_o, label = 'Training Accuracy')
plt.plot(neighbors, test_accuracy2_o, label = 'Testing New Data Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
# Create correlation matrix
corr_matrix = X_data.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.90
to_drop = [column for column in upper.columns if any(upper[column] > 0.90)]
print(to_drop)


X_data=X_data.drop(columns=to_drop)
print("\n Shape after highly correlated columns are deleted:",X_data.shape,"\t Dimesion reduced by:",np.subtract(X_data_o.shape, X_data.shape))
```

```python
# Build the model
rfc = RandomForestClassifier(random_state=0)
# Fit the model
rfc.fit(X_data, y_target)
# Print the results
#print("Features sorted by their score:")
#print(sorted(zip(map(lambda x: round(x, 4), rfc.feature_importances_), names), reverse=True))

importances = rfc.feature_importances_
indices = np.argsort(importances)

plt.figure(1)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices])
plt.xlabel('Relative Importance')

sfm = SelectFromModel(rfc, threshold=0.01)
sfm.fit(X_data, y_target)
```

```python
names = X_data.columns.values
RF_col = []
for feature_list_index in sfm.get_support(indices=True): 
    #print(names[feature_list_index])
    RF_col.append(names[feature_list_index])
print('\n Columns chosen using Random Forest:\n',RF_col)


X_data_RF=X_data[RF_col]
print('\n Shape of data after choosing important features:',X_data_RF.shape)
```

```python
#split data
X_data1_RF=X_data_RF[0:220]
y_target1=y_target[0:220]
X_data2_RF=X_data_RF[220:452]
y_target2= y_target[220:452]
```

```python
#knn using RF with few dimensions
X_train, X_test, y_train, y_test= train_test_split(X_data1_RF,y_target1,test_size=0.2, random_state=0)

# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 15)
train_accuracy = np.empty(len(neighbors))
#train_error = np.empty(len(neighbors))
test_accuracy= np.empty(len(neighbors))
#test_error = np.empty(len(neighbors))
test_accuracy2 = np.empty(len(neighbors))


# Loop over different values of k
for i, k in enumerate(neighbors):
    # Setup a k-NN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn.fit(X_train,y_train)
    
    
    #Compute accuracy on the training set
    
    train_accuracy[i] = knn.score(X_train, y_train)
    #train_error[i] =(1-knn.score(X_train, y_train))*100

    #Compute accuracy on the testing set
    knn.predict(X_test)
    test_accuracy[i] = knn.score(X_test, y_test)
    #test_error[i] =(1-knn.score(X_test, y_test))*100
    
    knn.predict(X_data2)
    test_accuracy2[i] = knn.score(X_data2_RF, y_target2)

# Generate plot
print('\n Training Accuracy \n',train_accuracy,'\n \n Testing Accuracy \n',test_accuracy,'\n \n Testing Accuracy on new data \n',test_accuracy2)
plt.title('k-NN: Varying Number of Neighbors with Random Forest')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy with RF')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy with RF')
plt.plot(neighbors, test_accuracy2, label = 'Testing New Data Accuracy with RF')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
